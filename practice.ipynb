```python
class node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def convert_to_linked_list(l):
    head = node()
    i = head

    for data in l:
        i.next = node(data)
        i = i.next
    return head.next

def print_lk(head):
    temp = head
    while temp:
        print(temp.val, end=" ->")
        temp = temp.next
    print("end")


def add_two_nums(l1, l2):
    head = node()
    temp, carry = head, 0

    while l1 or l2 or carry:
        x = l1.val if l1 else 0
        y = l2.val if l2 else 0
        total = x+y+carry

        carry = total//10
        temp.next = node(total%10)
        temp = temp.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    return head.next

l1 = convert_to_linked_list([2,4,3])
l2 = convert_to_linked_list([5,6,4])

print_lk(l1)
print_lk(l2)
r = add_two_nums(l1,l2)
print_lk(r)
```

```python
def merge(l1,l2):
    head = node()
    temp = head

    while l1 and l2:
        if l1.val <= l2.val:
            temp.next = l1
            l1 = l1.next
        else:
            temp.next = l2
            l2 = l2.next
        temp = temp.next
    temp.next = l1 if l1 else l2
    return head.next
merge(l1,l2)
```

```python
def remove_dupes(nums):
    if not nums:
        return 0
    if len(nums) == 1:
        return 1
    k = 1
    for i in range(1,len(nums)):
        if nums[i] != nums[i-1]:
            nums[k] = nums[i]
            k += 1
    print(nums)
    return k
remove_dupes([0,0,1,1,1,2,2,3,3,4])
```

```python
def removeElement(nums, val: int) -> int:
    k = 0
    for i in range(0, len(nums)):
        if nums[i] != val:
            if k!=i:
                nums[k] = nums[i]
            k+=1
    print(nums)
    return k
removeElement([3,2,2,3], 3)
```

```python
n = 1
def pascal(n):
    ans = [[1], [1,1]]
    if n == 1 or n == 2: return ans[0:n]
    for k in range(n-2):
        ref = ans[-1] # pick last item to calculate the upcoming value
        temp = [1] # pascal triangle starts with 1
        
        for i in range(0, len(ref)-1):
            temp.append(ref[i]+ref[i+1])
        temp.append(1)
        ans.append(temp)
        # temp = []
    return ans

res = pascal(n)
res
# ans = [[1], [1,1]]
# ans[0:2]
```

```python
rowIndex = 3
def new_pascal(rowIndex):
    ans = [[1], [1,1]]
    if rowIndex == 0 or rowIndex == 1: return ans[rowIndex]
    n = rowIndex # rowIndex - 1
    while n > 0:
        ref = ans[-1]
        temp = [1]
        for i in range(0, len(ref)-1):
            temp.append(ref[i] + ref[i+1])
        temp.append(1)
        ans.append(temp)
        n -= 1
    return ans[rowIndex] # [-1]
res = new_pascal(rowIndex)
res
```
